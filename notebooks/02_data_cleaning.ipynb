```python
import pandas as pd

df = pd.read_csv("../data/raw/results.csv")

df["date"] = pd.to_datetime(df["date"])
```

```python
print(df.shape)
```

```python
worldcup2026_df = df[
    (df["tournament"] == "FIFA World Cup")
    &
    (df["date"].dt.year == 2026)
].copy()

worldcup2026_df.shape
```

```python
historical_df = df[
    ~(
        (df["tournament"] == "FIFA World Cup")
        &
        (df["date"].dt.year == 2026)
    )
].copy()

historical_df.shape
```

```python
historical_df.isnull().sum()
```

```python
historical_df["home_score"] = historical_df["home_score"].astype(int)
historical_df["away_score"] = historical_df["away_score"].astype(int)
```

```python
historical_df.dtypes
```

```python
print("Antes:", historical_df.shape)

historical_df = historical_df.drop_duplicates(
    subset=[
        "date",
        "home_team",
        "away_team",
        "home_score",
        "away_score"
    ]
).reset_index(drop=True)

print("Después:", historical_df.shape)
```

```python
historical_df.to_csv(
    "../data/interim/historical_matches.csv",
    index=False
)

worldcup2026_df.to_csv(
    "../data/interim/worldcup2026_matches.csv",
    index=False
)
```
