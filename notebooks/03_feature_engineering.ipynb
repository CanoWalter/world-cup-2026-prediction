```python
import pandas as pd
import numpy as np

historical_df = pd.read_csv(
    "../data/interim/historical_matches.csv"
)

historical_df["date"] = pd.to_datetime(
    historical_df["date"]
)
```

```python
historical_df["result"] = np.where(
    historical_df["home_score"] > historical_df["away_score"],
    "H",
    np.where(
        historical_df["home_score"] < historical_df["away_score"],
        "A",
        "D"
    )
)
```

```python
historical_df["goal_difference"] = (
    historical_df["home_score"]
    - historical_df["away_score"]
)
```

```python
historical_df["total_goals"] = (
    historical_df["home_score"]
    + historical_df["away_score"]
)
```
