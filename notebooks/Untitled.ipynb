```python
import pandas as pd
import numpy as np

historical_df = pd.read_csv(
    "../data/interim/historical_matches.csv"
)

historical_df["date"] = pd.to_datetime(
    historical_df["date"]
)

worldcup_df = pd.read_csv(
    "../data/interim/worldcup2026_matches.csv"
)

worldcup_df["date"] = pd.to_datetime(
    worldcup_df["date"]
)
```

```python
worldcup_played = historical_df[
    historical_df["tournament"] == "FIFA World Cup"
].copy()

worldcup_played.tail()
```

```python
print(
    "Partidos mundialistas jugados:",
    len(worldcup_played)
)
```

```python
home_table = pd.DataFrame({

    "team": worldcup_played["home_team"],

    "gf": worldcup_played["home_score"],

    "ga": worldcup_played["away_score"],

    "points": np.where(
        worldcup_played["home_score"] >
        worldcup_played["away_score"],
        3,
        np.where(
            worldcup_played["home_score"] ==
            worldcup_played["away_score"],
            1,
            0
        )
    )
})
```

```python
away_table = pd.DataFrame({

    "team": worldcup_played["away_team"],

    "gf": worldcup_played["away_score"],

    "ga": worldcup_played["home_score"],

    "points": np.where(
        worldcup_played["away_score"] >
        worldcup_played["home_score"],
        3,
        np.where(
            worldcup_played["away_score"] ==
            worldcup_played["home_score"],
            1,
            0
        )
    )
})
```

```python
table = pd.concat([
    home_table,
    away_table
])

table = (
    table
    .groupby("team")
    .agg({
        "points":"sum",
        "gf":"sum",
        "ga":"sum"
    })
)

table["gd"] = (
    table["gf"]
    - table["ga"]
)

table = table.sort_values(
    ["points","gd","gf"],
    ascending=False
)

table.head(20)
```
